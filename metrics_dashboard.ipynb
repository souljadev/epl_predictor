# EPL Agent Metrics Dashboard
This notebook visualizes metrics over time from `models/history/`.

```
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

hist = Path('../models/history')
files = sorted(hist.glob('eval_*.csv'))
all_rows = []
for f in files:
    df = pd.read_csv(f)
    df['eval_file'] = f.name
    # Compute basic metrics
    df['total_actual'] = df['FTHG'] + df['FTAG']
    df['total_pred'] = df['ExpHomeGoals'] + df['ExpAwayGoals']
    df['abs_err_total'] = (df['total_actual'] - df['total_pred']).abs()
    all_rows.append(df)

if all_rows:
    full = pd.concat(all_rows)
    summary = full.groupby('eval_file')[['abs_err_total']].mean()
    summary.plot(kind='bar', title='Mean Absolute Error (Total Goals)')
    plt.show()
else:
    print('No evaluation files found.')
```
